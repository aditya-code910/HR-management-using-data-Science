```python
for col in df.columns:
    if df[col].dtype == 'object' or df[col].dtype == 'string':
        try:
            df[col] = df[col].astype(str)
            df[col] = df[col].str.replace('RM', '', regex=False)
            df[col] = pd.to_numeric(df[col], errors='coerce')
        except Exception as e:
            pass  # skip if column can't be cleaned this way
```

```python
df_bus = df.groupby('BusinessTravel').size().reset_index(name = 'Travel')

df_edu = df.groupby('EducationField').size().reset_index(name = 'education')

df_daily = df.groupby('DailyRate').size().reset_index(name = 'Daily')

df_Performance = df.groupby('PerformanceRating').size().reset_index(name = 'Performance')

df_salary = df.groupby('MonthlyIncome').size().reset_index(name = 'Salary')

attributes = [df_bus, df_edu, df_daily, df_Performance, df_salary]
```

```python
for attribute in attributes:
    plt.figure(figsize = (10, 5))
    plt.xlabel(str(attribute.columns[0]))
    plt.ylabel(str(attribute.columns[1]))
    plt.title('Exiting Employees')
    x = attribute[attribute.columns[0]]
    y = attribute[attribute.columns[1]]
    plt.bar(x, y)
    plt.grid(which = 'both', linestyle = ':')
```

```python
sns.displot(df['EmployeeNumber'], kde = True, bins =4 )
```

```python
plt.figure(figsize = (8, 10))
for attribute in attributes:
    total_exiting_employees = 1480
    high = attribute[attribute.columns[1]].max()
    percentage = round((high / total_exiting_employees)* 100, 2)
    name = attribute.columns[1]
    plt.ylabel(str(attribute.columns[1]))
    plt.title('Exiting Employees')
    plt.bar(name, percentage)
    print(str(name) + " " + str(percentage))
    plt.grid(which = 'both', linestyle = ':')
```

```python
df_salary_dept = df.groupby('Department')['MonthlyIncome'].mean().reset_index()

df_salary_dept = df_salary_dept.sort_values('MonthlyIncome', ascending=True)

plt.figure(figsize=(10, 6))
sns.barplot(data=df_salary_dept, x='MonthlyIncome', y='Department', palette='viridis')
plt.xlabel('Average Monthly Income')
plt.ylabel('Department')
plt.title('Average Salary by Department')
plt.tight_layout()
plt.show()
```

```python
edu = df['EducationField'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(edu, labels=edu.index, autopct='%1.1f%%', startangle=140)
plt.title('Employee Distribution by Education Field')
plt.axis('equal')  # Keeps pie chart circular
plt.show()
```

```python
df['Gender_bin'] = df['Gender'].map({'Female': 1, 'Male': 0})

gender_percent = df.groupby('EducationField')['Gender_bin'].mean().reset_index()

plt.figure(figsize=(10, 8))
sns.barplot(data=gender_percent, x='EducationField', y='Gender_bin', palette='mako')
plt.title('Proportion of Females in Each Education Field')
plt.xlabel('Education Field')
plt.ylabel('Proportion Female')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
