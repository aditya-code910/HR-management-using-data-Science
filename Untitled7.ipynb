```python
import pandas as pd
from datetime import datetime, time

# Load the Excel file
df = pd.read_excel("time.xlsx")

# Define shift start time
shift_start = time(9, 30, 0)  # 9:30 AM

# Identify all 'time in' columns
time_in_cols = [col for col in df.columns if 'time in' in str(col).lower()]

# Create result dataframe
result = pd.DataFrame()
result['empid'] = df['empid']
result['name'] = df['name']

# Process each day's time in
for idx, col in enumerate(time_in_cols):
    # Get corresponding date
    date_col_idx = df.columns.get_loc(col) - 1
    date = df.columns[date_col_idx] if isinstance(df.columns[date_col_idx], pd.Timestamp) else f"Day {idx+1}"

    # Updated status function with time in output
    def get_status(val):
        try:
            in_time = datetime.strptime(val.strip(), "%H:%M:%S").time()
            formatted_time = in_time.strftime("%H:%M")
            shift_dt = datetime.combine(datetime.today(), shift_start)
            in_dt = datetime.combine(datetime.today(), in_time)
            grace_period = shift_dt + timedelta(minutes=10)

            if in_dt < shift_dt:
              return f"Early ({formatted_time})"
            elif shift_dt <= in_dt <= grace_period:
              return f"On Time ({formatted_time})"
            else:
              return f"Late ({formatted_time})"
        except:
          return 'Absent/Invalid'


    result[str(date.date() if isinstance(date, pd.Timestamp) else date)] = df[col].astype(str).apply(get_status)

result.head(50)
```

```python
import pandas as pd
from datetime import datetime

# Load your data
df = pd.read_excel("time1.xlsx")

# List of Time In and Time Out column pairs
time_pairs = [(f'Day {i} In', f'Day {i} Out') for i in range(1, 23)]  # assuming 22 days

# Function to calculate hours between in and out
def calculate_hours(in_time, out_time):
    try:
        in_dt = pd.to_datetime(str(in_time)).time()
        out_dt = pd.to_datetime(str(out_time)).time()
        in_dt = datetime.combine(datetime.today(), in_dt)
        out_dt = datetime.combine(datetime.today(), out_dt)
        duration = (out_dt - in_dt).seconds / 3600
        return duration if duration >= 0 else 0
    except:
        return 0

# Count overtime days per employee
df['overtime_days'] = 0

for in_col, out_col in time_pairs:
    if in_col in df.columns and out_col in df.columns:
        overtime_hours = df.apply(lambda row: calculate_hours(row[in_col], row[out_col]), axis=1)
        df['overtime_days'] += (overtime_hours > 8).astype(int)

# Result: show employees with any overtime days
overtime_df = df[df['overtime_days'] > 0][['empid', 'name', 'overtime_days']]
print("Employees with Overtime:\n")
print(overtime_df)
```

```python
import pandas as pd

working_hour_cols = [col for col in df.columns if 'total working hour' in str(col).lower()]

def count_overtime_days(row):
    count = 0
    for col in working_hour_cols:
        val = row[col]
        if pd.notna(val):
            try:
                if isinstance(val, pd._libs.tslibs.nattype.NaTType):
                    continue
                elif isinstance(val, pd.Timedelta):
                    hours = val.total_seconds() / 3600
                elif isinstance(val, str):
                    t = pd.to_timedelta(val)
                    hours = t.total_seconds() / 3600
                else:
                    hours = float(val.hour) + float(val.minute)/60 if hasattr(val, 'hour') else float(val)

                if hours > 8:
                    count += 1
            except Exception:
                continue
    return count


df['overtime_days'] = df.apply(count_overtime_days, axis=1)

overtime_df = df[df['overtime_days'] > 0][['empid', 'name', 'overtime_days']]

print("âœ… Employees with Overtime Days:\n")
print(overtime_df)
```

```python
import pandas as pd

# Make sure all column names are strings
df.columns = df.columns.map(str)

# Identify columns containing total working hour info
working_hour_cols = [col for col in df.columns if 'total working hour' in col.lower()]

# Helper to convert to hours
def convert_to_hours(val):
    if pd.isnull(val):
        return 0
    if isinstance(val, str):
        try:
            t = pd.to_datetime(val).time()
        except:
            return 0
    elif isinstance(val, pd.Timestamp):
        t = val.time()
    elif isinstance(val, datetime):
        t = val.time()
    else:
        t = val  # Assuming already time object
    return t.hour + t.minute / 60 + t.second / 3600

# Calculate total working hours per employee
df['total_working_hours'] = df[working_hour_cols].applymap(convert_to_hours).sum(axis=1)

# Count how many days they worked overtime (>8 hours)
df['overtime_days'] = df[working_hour_cols].applymap(lambda x: convert_to_hours(x) > 8).sum(axis=1)

# Get top 30
top_30 = df.sort_values(by='total_working_hours', ascending=False).head(30)

# Select useful columns
result = top_30[['empid', 'name', 'total_working_hours', 'overtime_days']]

# Output result
print("Top 30 Employees with Highest Total Working Hours & Overtime Days:\n")
print(result)

# Export to Excel
result.to_excel("top_30_overtime_employees.xlsx", index=False)
```
